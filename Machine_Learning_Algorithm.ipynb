### **1. Linear Regression Algorithm**

```python
# Importing packages

import pandas as pd
import numpy as np
import sklearn
```

```python

data = pd.read_csv('/content/housing.csv')
data.head()
```

```python
data.describe()
```

```python
# Processing the data (Removing NaN values): Let's check if there is any null value
data.isna().sum()
```

```python
# There are no missing values in the housing data
data.describe()
```

Splitting the model data with 70% for the training. 30% for the testing





```python
from pandas.core.arrays.sparse import array
from sklearn.model_selection import train_test_split
X = np.array(data.iloc[:,0:13])
Y = np.array(data["MEDV"])

# Testing the data size is of 30% of the entire data
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.30, random_state=5)
```

**Using Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

# Load our 1st model
lr = LinearRegression()
# train the model on training data

lr.fit(x_train, y_train)
#Predict the testing data so that we can later evaluate the model
pred_lr = lr.predict(x_test)

```

**Model Evaluation**

```python
# Error for linear regression
mse_lr = sklearn.metrics.mean_squared_error(y_test, pred_lr, squared = False)
print("Error for Linear Regression = {}" .format(mse_lr))
```

### **2. KNN Algorithm**

- Input data is indexed to find the closest neighbors
- Data is compared in the referencing pgase to save time
- belongs to the category of lazy learns

```python
# Load the KNN Model
Nn = KNeighborsRegressor(3)
# Train the model
Nn.fit(x_train, y_train)
pred_Nn = Nn.predict(x_test)
```

Hyperparameter Tuning

```python
import sklearn
for i in range(1,50):
  model = KNeighborsRegressor(i)
  model.fit(x_train, y_train)
  pred_y = model.predict(x_test)
  mse = sklearn.metrics.mean_squared_error(y_test,pred_y, squared = False)
  print("{}  Error for k = {} ".format(mse,i))
```

Model Evaluation

```python
# Error for KNN Algorithm

mse_Nn = sklearn.metrics.mean_squared_error(y_test, pred_Nn, squared =False)
print("Error for KNN = {} ".format(mse_Nn))
```

### **3. Support Vector Machines (SVMs)**

- Considered as a black box techniques due to presence of unknown parameters
- Data separation for non-linear classes are done using kernel tricks
- Parameters are tuned using kernel functions

Important aspects of ML:
 - Very flexible working with variety  of data (Unstructured, structured and semi-structured)
 - Overfitting is very less compared to other models
 - Very popular in healthcare and banking sectors

```python
# Importing all the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
social = pd.read_csv('/content/Social_Network_Ads.csv')
social.head(50)
```

Filtering out columns to retain age and salary columns

```python

X =  pd.DataFrame(social['EstimatedSalary'], columns=['EstimatedSalary'])
y= pd.DataFrame(social['Purchased'], columns=['Purchased'])

```

```python
X.head(15)
```

```python
y.head(15)
```

Splitting data for training and testing (75/25)

```python
# testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.25, random_state=0)
```

Scaling using Standard Scaler for Normal Distibition

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Building the model using RBF Kernel

```python
from sklearn.svm import SVC
classifier_rbf = SVC(kernel='rbf', random_state = 0)
classifier_rbf.fit(X_train, y_train)
y_pred_rbf = classifier_rbf.predict(X_test)
```

Priting the confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm_rbf =  confusion_matrix(y_test, y_pred_rbf)
print(cm_rbf)
```

Generating classification Report

```python
from sklearn.metrics import classification_report
class_report_rbf = classification_report(y_test, y_pred_rbf)

print(class_report_rbf)
```

From the above results,

**Precision:**

Precision measures how many of the predicted positive instances (class 1, "Not Purchased") were actually correct.
- For class 1 (Purchased), the precision is 0.80, which means that out of all instances predicted as "Purchased," 80% were correct.
- For class 0 (Not Purchased), the precision is 0.88, which means that out of all instances predicted as "Not Purchased," 88% were correct.

**Recall:**

Recall (also known as Sensitivity or True Positive Rate) measures how many of the actual positive instances (class 1, "Not Purchased") were correctly predicted by the model.
- For class 1 (Purchased), the recall is 0.97, which means that the model correctly identified 97% of all actual "Purchased" instances.

For class 0 (Not Purchased), the recall is 0.47, which means that the model correctly identified 47% of all actual "Not Purchased" instances.

**F1-Score:**

The F1-score is the harmonic mean of precision and recall. It provides a balance between the two metrics and is especially useful when dealing with imbalanced datasets.
- For class 1 (Purchased), the F1-score is 0.87.
For class 0 (Not Purchased), the - F1-score is 0.61.

**Support:**

Support represents the number of instances in each class in your dataset.

There are 68 instances of class 1 (Purchased) and 32 instances of class 0 (Not Purchased).

**Accuracy:**

The overall accuracy of the model is 0.81, which means that it correctly classified 81% of all instances in the dataset.

**Macro Average:**

The macro average calculates the metrics independently for each class and then takes the average. It gives equal weight to each class.
The macro average precision is 0.84, recall is 0.72, and F1-score is 0.74.

**Weighted Average:**

The weighted average also calculates the metrics for each class, but it gives more weight to the class with a larger number of instances. It provides a better overall assessment when classes are imbalanced.

The weighted average precision is 0.82, recall is 0.81, and F1-score is 0.79.
